```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('Dataset.xlsx')
```

```python
df.head()
```

```python
# 1.finding the maximum sold product in each groupement 
```

```python
# finding all the unique groups in the dataset
```

```python
print(np.unique(df[['Groupement']].values))
```

```python
# dividing all unique groupement in different dataframes
```

```python
C10=df.loc[df['Groupement'] == "C10"]
```

```python
C10.head()
```

```python
# finding which product id has been sold in maximum quantity
```

```python
MAX_C10=C10.groupby('Material')['Order qty'].sum().reset_index()
```

```python
print(MAX_C10)
```

```python
# finding the product id of the maximum sold product
```

```python
column = MAX_C10["Order qty"]
max_value = column.max()
MAX_C10.loc[MAX_C10['Order qty'] == max_value]
```

```python
# creating a new dataframe where we will have all those wholesalers who have not bought the maximum sold product
```

```python
C10_with_no_max=C10.loc[C10['Material'] != 10949]
```

```python
# converting that dataframe into an excel file
```

```python
writer = pd.ExcelWriter('C10_with_no_max.xlsx')
C10_with_no_max.to_excel(writer)
writer.save()
```

```python
# 2.finding which product id has maximum frequency of purchase
```

```python
# creating a dataframe dff which depicts which product id has maximum frequency of purchase; this is done by summing the delivery flag
```

```python
dff=df.groupby('Material')['delivery_flag'].sum().reset_index()
```

```python
print(dff)
```

```python
# finding which product id has maximum frequency
```

```python
column = dff["delivery_flag"]
max_value = column.max()
dff.loc[dff['delivery_flag'] == max_value]
```

```python
# finding all wholesalers who have not bought maximum frequency product
```

```python
frequency_based_output=df.loc[df['Material'] != 11577]
```

```python
frequency_based_output.head()
```

```python
# converting this data into excel sheet
```

```python
writer = pd.ExcelWriter('frequency_based.xlsx')
frequency_based_output.to_excel(writer)
writer.save()
```
